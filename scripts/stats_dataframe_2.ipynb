```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
# removed first row and column
# df = pd.read_csv ("/Users/kellyfang/Downloads/11.154/separation_data_set.csv")
df = pd.read_csv ("separation_data_set.csv")
df["lat"] = ""
df["lon"] = ""

display(df)
```

```python
from datetime import datetime

def days_between(d1, d2):
    d1 = datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.strptime(d2, "%Y-%m-%d")
    return abs((d2 - d1).days)


start = df["DATE_APPREHENDED"]
end = df["Date Discharged"]


duration_col = []

for i in range(len(df)):
    d1 = start[i]
    d2 = end[i]
    
    if type(d2) == float or len(d2)<5:
        duration_col.append(0)
    else:
        split1 = d1.split("/")
       
        date1 = split1[2] + "-" + split1[0]+"-" + split1[1]
        

        split2 = d2.split("/")
        date2 = split2[2] + "-" + split2[0]+"-" + split2[1]
    
        dur = days_between(date1,date2)
        
        duration_col.append(dur)
        
df_new = df.assign(Duration = duration_col)
display(df_new)
```

```python
h = df_new[["FACILITY_APPROVED", "UAC STATUS"]].copy()
discharge_binary = []
for i in h["UAC STATUS"]:
    if i == "DISCHARGED":
        discharge_binary.append(1)
    else:
        discharge_binary.append(0)
        
df_discharge = pd.DataFrame({'discharge_rate':discharge_binary})

df_discharge_rate = h.join(df_discharge).groupby(["FACILITY_APPROVED"]).mean()
```

```python
df_n = df_new[["FACILITY_APPROVED", "lat", "lon","Duration"]].copy()
display(df_n)
```

```python
df_avg_duration = df_n.groupby(["FACILITY_APPROVED"]).mean()
```

```python
stats_df = df_avg_duration.join(df_discharge_rate)
display(stats_df)
```

```python
stats_df.to_csv('../facility_stats.csv')
```
