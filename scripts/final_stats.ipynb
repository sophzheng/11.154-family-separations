```python
from datetime import datetime

def days_between(d1, d2):
    d1 = datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.strptime(d2, "%Y-%m-%d")
    return abs((d2 - d1).days)


start = df["DATE_APPREHENDED"]
end = df["Date Discharged"]


duration_col = []

for i in range(len(df)):
    d1 = start[i]
    d2 = end[i]
    
    if type(d2) == float or len(d2)<5:
        duration_col.append(0)
    else:
        split1 = d1.split("/")
       
        date1 = split1[2] + "-" + split1[0]+"-" + split1[1]
        

        split2 = d2.split("/")
        date2 = split2[2] + "-" + split2[0]+"-" + split2[1]
    
        dur = days_between(date1,date2)
        
        duration_col.append(dur)
        
df_new = df.assign(Duration = duration_col)
display(df_new)
```

```python
# calculate the avg durationa across ALL children:
df_avg = df_new.loc[(df_new["Duration"] != 0)]#.reset_index()
#df_avg.drop(['index'])
display(df_avg)
df_avg["Duration"].mean()
```
